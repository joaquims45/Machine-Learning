```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## **Carga de dataframe**


```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

## **Cambiar total_charges a numerico**

```python
df.info()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
```

```python
df.info()
```

## **Manejo de datos nulos**

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head(5)
```

```python
df.drop('customerID', axis=1, inplace=True)
```

```python
## **Convertir variable objetivo a numérico**
```

```python
df['Churn'].replace(to_replace='Yes', value= 1, inplace=True)
df['Churn'].replace(to_replace='No', value= 0, inplace=True)
```

```python
df['Churn']
```

```python
df_processing = df.copy()
```

## **Manejo de variables categoricas**

```python
df_processing = pd.get_dummies(df_processing)
df_processing.head(5)
```

## **Analisis de correlación**

```python
fig = plt.figure(figsize=(15,9))
df_processing.corr()['Churn'].sort_values(ascending=True).plot(kind='bar')
plt.show()
```

## **Escalabilidad de datos**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df_processing_scaled = scaler.fit_transform(df_processing)
df_processing_scaled = pd.DataFrame(df_processing_scaled)

df_processing_scaled.columns = df_processing.columns

df_processing_scaled.head(5)
```

```python
import seaborn as sns
```

```python
df['gender'] = df['gender'].astype(str)
df['Churn'] = df['Churn'].astype(str)
sns.countplot(data=df, x='gender', hue='Churn')
plt.show()
```

```python
def plot_categorial(column):
    fig = plt.figure(figsize=(10,10))
    sns.countplot(data=df, x=column,hue='Churn')
    plt.show()
```

```python
column_cat = df.select_dtypes(include='object').columns
```

```python
for _ in column_cat:
    plot_categorial(_)
```

```python
fig = plt.figure(figsize=(10,10))
sns.pairplot(data=df, hue='Churn')
plt.show()
```

## **Escalabilidad de datos**

```python
X = df_processing_scaled.drop('Churn',axis=1)
y = df_processing_scaled['Churn'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn import metrics

prediction_test = model.predict(X_test)
print(metrics.accuracy_score(y_test, prediction_test))
```

## **Evaluacion del modelo**

```python
model.predict_proba(X_test)
```

```python
model.coef_
```

```python
model.feature_names_in_
```

```python
weights = pd.Series(model.coef_[0],
                   index=X.columns.values)
print(weights.sort_values(ascending=False)[:10].plot(kind='bar'))
```

```python
print(weights.sort_values(ascending=False)[-10:].plot(kind='bar'))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, prediction_test, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=model.classes_)
disp.plot(cmap='gray')
plt.show()
```
